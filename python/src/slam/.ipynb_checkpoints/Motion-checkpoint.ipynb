```python
import matplotlib.pylab as plt
from mpl_toolkits import mplot3d
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('slam.csv')
```

```python
data.head()
```

```python
x = data['X']
y = data['Y']
z = data['Z']
```

```python
plt.plot(x, z)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(x, y, z, 'gray')
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(x, y, z, 'gray')
```
